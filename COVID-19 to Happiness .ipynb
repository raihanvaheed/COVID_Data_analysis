```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing data
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
#Delete usless columns
df = corona_dataset_csv.drop(["Lat", "Long"], axis=1)
df.head(10)
```

```python
# Aggregating country data
corona_dataset_aggregated = df.groupby('Country/Region').sum()
corona_dataset_aggregated.head(10)
```

```python
#Checking Data
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()

plt.legend()
```

```python
#First 3 days
corona_dataset_aggregated.loc['China'][:3].plot()
```

```python
# Differenttiate to find highest infection rate
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#Maximum infection rate 
def max_Infection_rate(Country):
    
    print("The maximum increase of infections in one day in {} was {}".format(Country, str(corona_dataset_aggregated.loc[Country].diff().max())))
```

```python
max_Infection_rate('China')
max_Infection_rate('Italy')
max_Infection_rate('Spain')
max_Infection_rate('United Arab Emirates')
```

```python
# Add  max infection rate
Countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for Country in Countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[Country].diff().max())
corona_dataset_aggregated['max_infection_rate'] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
# New collection of data
corona_data = pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
corona_data.head()
```

```python
# Load Happiness report data
happiness_report_csv = pd.read_csv("Happiness_report.csv")
happiness_report_csv.head()
```

```python
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report_csv.drop(useless_cols, inplace=True, axis=1)
happiness_report_csv.head()
```

```python
happiness_report_csv.set_index('Country or region', inplace=True)
happiness_report_csv.head()
```

```python
corona_data.shape
```

```python
happiness_report_csv.shape
```

```python
data = corona_data.join(happiness_report_csv, how="inner")
data.head(10)
```

```python
# Make a correlation Matrix
data.corr()
# There is a positive correlation with each factor (O.O wow)
```

```python
def plotter(x, y, log_y, type_graph):
    X = data[x]
    if log_y == True:
        Y = np.log(data[y])
    else:
        Y = data[y]
    if type_graph == 'scatter':
        sns.scatterplot(X, Y)
    elif type_graph == 'reg':
        sns.regplot(X, Y)
```

```python
# Visualise results
# GDP vs max infection rate
x = data['GDP per capita']
y = data['max_infection_rate']
sns.scatterplot(x, np.log(y)) # Using log scaling in Y axis because of large values
```

```python
sns.regplot(x, np.log(y))
```

```python
# Social Support vs Maximum infection rate
plotter('Social support', 'max_infection_rate', True, 'scatter')
```

```python
plotter('Social support', 'max_infection_rate', True, 'reg')
```

```python
# Healthy Life expectancy vs max infection rate
plotter('Healthy life expectancy', 'max_infection_rate', True, 'scatter')
```
